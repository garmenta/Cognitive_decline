```python
# CORRELATIONS OF COGNITIVE STATUS AS A BAR PLOT

# Define the thematic DataFrames
demographic = data[demographic_vars]
behavioral = data[behavioral_vars]
economic = data[economic_vars]
health = data[health_vars]
psychological_and_mental = data[psychological_and_mental_vars]
social = data[social_vars]

# Define a function to calculate correlations and return them sorted by absolute value
def calculate_and_sort_correlations(df, target_column):
    correlations = df.corr()[target_column].drop(target_column, errors='ignore')  # Exclude self-correlation
    abs_correlations = correlations.abs().sort_values(ascending=False)
    sorted_correlations = correlations.loc[abs_correlations.index]  # Apply sorted order based on absolute values
    return sorted_correlations

# Lists of DataFrames and their corresponding names for iteration
dataframes = [demographic, behavioral, economic, health, psychological_and_mental, social]
dataframe_names = ["Demographic Variables", "Behavioral Variables", "Economic Variables", "Health-Related Variables", "Psychological and Mntal Variables", "Social Variables"]

# Process each DataFrame
for df, name in zip(dataframes, dataframe_names):
    sorted_correlations = calculate_and_sort_correlations(df, 'COGNTVsta')
    
    plt.figure(figsize=(8, 6))
    bar_colors = ['#ADD8E6' if value >= 0 else '#FFA07A' for value in sorted_correlations.values]  # Light blue or orange based on correlation sign
    bars = plt.bar(sorted_correlations.index, sorted_correlations.values, color=bar_colors)
    plt.xticks(rotation=90, fontsize=8)
    plt.yticks(fontsize=8)
    plt.ylim(-0.2,0.4)
    plt.title(f"Correlations between Cognitive Status and {name}")
    plt.xlabel('Variable name')
    plt.ylabel('Correlations')
    plt.tight_layout()

    # Add vertically aligned, smaller-sized text on top of each bar
    for bar, corr in zip(bars, sorted_correlations.round(3)):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height + 0.01, corr, ha='center', va='bottom', rotation=90, fontsize=7)
     
    # Save the plot
    # Before displaying the figure, save it
    plot_path = os.path.expanduser(f'~/Documents/NYCDSA/Python project/{name}_correlations.png')
    plt.savefig(plot_path)
    
    # Now display the figure in the notebook
    plt.show()
    
    # After showing, close the plot to free memory and avoid conflicts with further plots
    plt.close()
```

```python
def calculate_and_sort_correlations(df):
    correlations = df.corr().abs().stack().reset_index()
    correlations.columns = ['variable 1', 'variable 2', 'correlation']
    correlations_clean = correlations[(correlations['variable 1'] != correlations['variable 2']) & (correlations['correlation']>=.5)]

 # Sort the correlations from highest to lowest
    sorted_correlations = correlations_clean.sort_values(by='correlation', ascending=False).reset_index(drop=True)
    
    return sorted_correlations

correlations_above_50percent = calculate_and_sort_correlations(data_clean)
print(correlations_above_50percent)
correlations_above_50percent.to_csv('correlations_above_50percent.csv', index=False)
```

```python
from scipy.stats import chi2_contingency

# Custom labels for the y-axis
labels = {
    0: "0.Normal",
    1: "1.Normal with\n instrumental\n impairment",
    2: "2.Cognitive\n impairment\n no dementia",
    3: "3.Dementia"
}

# Custom labels for the y-axis
labels_bmi = {
    1.0: "Underweight",
    2.0: "Normal",
    3.0: "Overweight",
    4.0: "Obese"
}

# Map the 'COGNTVsta' values to the custom labels for plotting
data['COGNTVsta_labels'] = data['COGNTVsta'].map(labels)
data['BMIcat2_labels'] = data['BMIcat2'].map(labels_bmi)
data['BMIcat2_labels'] = pd.Categorical(data['BMIcat2_labels'],
                                        categories=labels_bmi.values(),
                                        ordered=True)

contingency_table = pd.crosstab(data['BMIcat2_labels'], data['COGNTVsta_labels'])
chi2, p, dof, expected = chi2_contingency(contingency_table)

# PLOTTING OBSERVED FREQUENCIES
plt.figure(figsize=(5, 4))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='coolwarm', annot_kws={"fontsize":7})
plt.title('Observed Frequencies', fontsize=10)
plt.ylabel('BMI Category', fontsize=9)
plt.xlabel('Cognitive Status', fontsize=9)
plt.gca().invert_yaxis()
plt.yticks(fontsize=7, rotation=0) 
plt.xticks(rotation=90, fontsize=7) 
plt.tight_layout()
plt.savefig('Observed Frequencies.png', dpi=500)
plt.show()

# PLOTTING EXPECTED FREQUENCIES
plt.figure(figsize=(5, 4))

# Convert the 'expected' array to a DataFrame for easier plotting with labels
expected_df = pd.DataFrame(expected, 
                           index=contingency_table.index, 
                           columns=contingency_table.columns)

sns.heatmap(expected_df, annot=True, fmt='0.0f', cmap='coolwarm', annot_kws={"fontsize":7})
plt.title('Expected Frequencies under Independence', fontsize=10)
plt.ylabel('BMI Category', fontsize=9)
plt.xlabel('Cognitive Status', fontsize=9)
plt.xticks(rotation=90, fontsize=7) 
plt.yticks(rotation=0, fontsize=7)

# Here we use the DataFrame's index and columns to set yticks and xticks labels
plt.xticks(np.arange(len(expected_df.columns)) + 0.5, expected_df.columns, rotation=90, fontsize=7)
plt.yticks(np.arange(len(expected_df.index)) + 0.5, expected_df.index, rotation=0, fontsize=7)

plt.gca().invert_yaxis()  # This inverts the y-axis to match your table layout
plt.tight_layout()
plt.savefig('Expected Frequencies under Independence.png', dpi=500)
plt.show()
```

```python
labels = {
    0: "0.Normal",
    1: "1.Normal with\ninstrumental\nimpairment",  # Ensure the format matches your data exactly
    2: "2.Cognitive\nimpairment\nno dementia",  # Ensure the format matches your data exactly
    3: "3.Dementia"
}

# Convert 'COGNTVsta' codes to labels
data['COGNTVsta_labels'] = data['COGNTVsta'].map(labels)

# Ensure 'COGNTVsta_labels' is categorical with the specified categories and order
data['COGNTVsta_labels'] = pd.Categorical(data['COGNTVsta_labels'],
                                          categories=labels.values(),
                                          ordered=True)

# Ensure 'COGNTVsta_labels' is categorical with the specified categories and order
if not pd.api.types.is_categorical_dtype(data['COGNTVsta_labels']):
    data['COGNTVsta_labels'] = pd.Categorical(data['COGNTVsta_labels'],
                                              categories=labels.values(),
                                              ordered=True)

# Initializing a 2x2 subplot grid
fig, ax = plt.subplots(2, 2, figsize=(12, 10))  # Adjust figsize as needed

def annotate_medians(ax, data, x, y):
    medians = data.groupby(y)[x].median()
    categories = data[y].cat.categories
    for i, category in enumerate(categories):
        median_val = medians[category]
        # Check if median_val is finite
        if pd.notnull(median_val):  # Checks for NaN and inf/-inf
            ax.text(median_val, i, f'{median_val:.1f}', ha='center', va='center', 
                    fontweight='bold', color='black', fontsize=8)

# First Boxplot: Age
sns.boxplot(x=data['age'], y=data['COGNTVsta_labels'], ax=ax[0, 0], color=(0.4, 0.6, 0.8), orient='h')
ax[0, 0].set_title('Boxplot of Cognitive Status by Years of Age')
ax[0, 0].set_ylabel('Cognitive status')
ax[0, 0].tick_params(axis='y', labelsize=8)
ax[0, 0].set_xlabel('Years of age')
ax[0, 0].invert_yaxis()
annotate_medians(ax[0, 0], data, 'age', 'COGNTVsta_labels')

# Second Boxplot: Social Support
sns.boxplot(x=data['socialSPT_tot'], y=data['COGNTVsta_labels'], ax=ax[0, 1], color='salmon', orient='h')
ax[0, 1].set_title('Boxplot of Cognitive Status by Social Support')
ax[0, 1].set_ylabel('')
ax[0, 1].tick_params(axis='y', labelsize=8)
ax[0, 1].set_xlabel('Total social support')
ax[0, 1].invert_yaxis()
annotate_medians(ax[0, 1], data, 'socialSPT_tot', 'COGNTVsta_labels')

# Third Boxplot: Locus of Control
sns.boxplot(x=data['locusCNTRL'], y=data['COGNTVsta_labels'], ax=ax[1, 0], color='lightgreen', orient='h')  
ax[1, 0].set_title('Boxplot of Cognitive Status by Locus of Control')
ax[1, 0].set_ylabel('Cognitive status')
ax[1, 0].tick_params(axis='y', labelsize=8)
ax[1, 0].set_xlabel('Locus of Control')
ax[1, 0].invert_yaxis()

annotate_medians(ax[1, 0], data, 'locusCNTRL', 'COGNTVsta_labels')

# Fourth Boxplot: Years of Education
sns.boxplot(x=data['yrsEDU'], y=data['COGNTVsta_labels'], ax=ax[1, 1], color='lightgray', orient='h')  
ax[1, 1].set_title('Boxplot of Cognitive Status by Years of Education')
ax[1, 1].set_ylabel('')
ax[1, 1].tick_params(axis='y', labelsize=8)
ax[1, 1].set_xlabel('Years of education')
ax[1, 1].invert_yaxis()
annotate_medians(ax[1, 1], data, 'yrsEDU', 'COGNTVsta_labels')

# Adjust layout
plt.subplots_adjust(wspace=0.5, hspace=0.5)  # Adds gaps between subplots
plt.tight_layout()

plt.savefig('Boxplot of Cognitive Status Overview boxplots.png')  # Save one figure with all subplots
plt.show()
```
